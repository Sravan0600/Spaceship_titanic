```python
import pandas as pd
col_names = ['PassengerId', 'HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'Age', 'VIP', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'Name', 'Transported']
# load dataset
space_d = pd.read_csv("train.csv", header=None, names=col_names)
```

```python
space_d = space_d.drop(index=0)
```

```python
test_col_names = ['PassengerId', 'HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'Age', 'VIP', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'Name']
test_space_d = pd.read_csv("test.csv", header=None, names=test_col_names)
```

```python
test_space_d = test_space_d.drop(index=0)
```

```python
space_d = space_d.drop('Name', axis = 1)
test_space_d = test_space_d.drop('Name', axis =1)
```

```python
#summary of categorial data if there are any
cat_columns = [col for col in space_d.columns if space_d[col].dtype == 'object' or (space_d[col].dtype in ['int64', 'float65'] and space_d[col].nunique() <= 2)]
print(space_d[cat_columns].describe(include=['object', 'int64', 'float64']))
```

```python
#converting the required columns into float
columns_to_convert = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']

# Convert the specified columns to float
space_d[columns_to_convert] = space_d[columns_to_convert].astype(float)

print(space_d)
print(space_d.dtypes)
```

```python
#dong the same for test data_set
columns_to_convert = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']

# Convert the specified columns to float
test_space_d[columns_to_convert] = test_space_d[columns_to_convert].astype(float)

print(space_d)
print(space_d.dtypes)
```

```python
#dealing with the missing values for the required columns
print(space_d.isnull().sum())
```

```python
#home_planet column is object so using mode imputaion
object_cols = ['HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'VIP']
# Impute missing values with mode
for col in object_cols:
    mode = space_d[col].mode()[0]  # Find the mode
    space_d[col].fillna(mode, inplace=True)  # Replace missing values with mode

# Check if the missing values are replaced
print(space_d.isnull().sum())
```

```python
#now filling up the other columns of missing values as well whose data_type is float
float_columns = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
# Impute missing values with mean for each float column
for col in float_cols:
    mean_value = space_d[col].mean()  # Compute mean
    space_d[col].fillna(mean_value, inplace=True)  # Replace missing values with mean
# Check if the missing values are replaced
print(space_d.isnull().sum())

```

```python
space_d['Age'].fillna(mean_value, inplace = True)
print(space_d.isnull().sum())
```

```python
#similarily doing the same thing for test data_set
#dealing with the missing values for the required columns
print(test_space_d.isnull().sum())
```

```python

object_cols = ['HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'VIP']
# Impute missing values with mode
for col in object_cols:
    mode = test_space_d[col].mode()[0]  # Find the mode
    test_space_d[col].fillna(mode, inplace=True)  # Replace missing values with mode

# Check if the missing values are replaced
print(test_space_d.isnull().sum())
```

```python
#now filling up the other columns of missing values as well whose data_type is float
float_columns = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
# Impute missing values with mean for each float column
for col in float_cols:
    mean_value = test_space_d[col].mean()  # Compute mean
    test_space_d[col].fillna(mean_value, inplace=True)  # Replace missing values with mean
# Check if the missing values are replaced
print(test_space_d.isnull().sum())
```

```python
test_space_d['Age'].fillna(mean_value, inplace = True)
print(test_space_d.isnull().sum())
```

```python
#we have to handle categorical variables in our dataset now into integers
from sklearn.preprocessing import LabelEncoder
cat_cols_space_d = ['PassengerId', 'HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'VIP','Transported']
# Initialize LabelEncoder
label_encoder = LabelEncoder()
# Encode each categorical column
for col in cat_cols_space_d:
    if space_d[col].dtype == 'object':  # Ensure the column is of type object
        space_d[col] = label_encoder.fit_transform(space_d[col].astype(str))

# Check the result
print(space_d.head())
print(space_d.dtypes)
```

```python
# Convert numeric columns to float
numeric_columns = space_d.select_dtypes(include=['int64', 'float64']).columns
space_d[numeric_columns] = space_d[numeric_columns].astype(float)
```

```python
space_d.dtypes
```

```python
#similarily handling categorical values in test data_set
from sklearn.preprocessing import LabelEncoder
cat_cols_test_space_d = ['PassengerId', 'HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'VIP']
# Initialize LabelEncoder
label_encoder = LabelEncoder()
# Encode each categorical column
for col in cat_cols_test_space_d:
    if test_space_d[col].dtype == 'object':  # Ensure the column is of type object
        test_space_d[col] = label_encoder.fit_transform(test_space_d[col].astype(str))

# Check the result
print(test_space_d.head())
print(test_space_d.dtypes)
```

```python
# Convert numeric columns to float
numeric_columns = test_space_d.select_dtypes(include=['int64', 'float64']).columns
test_space_d[numeric_columns] = test_space_d[numeric_columns].astype(float)
test_space_d.dtypes
```

```python
#now building a corelation matrix to see what features are linked with transported column in train data
#correlation matrix
numerical_cols = ['PassengerId', 'HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'Age', 'VIP', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'Transported']
corr_matrix = space_d[numerical_cols].corr()
print(corr_matrix)
#visualization of correaltion matrix
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show
```

```python
#now for selecting features that are in range of -0.5 to 0.5
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier  



# Define features and target variable for training
features = ['VRDeck', 'Spa', 'RoomService', 'Destination', 'CryoSleep', 'HomePlanet']
x_train = space_d[features]
y_train = space_d['Transported']

# Split data into training and validation sets
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)

# Initialize your model (example with RandomForestClassifier)
model = RandomForestClassifier(random_state=42)

# Train the model on the training data
model.fit(x_train, y_train)

# Evaluate model performance on the validation set
accuracy = model.score(x_val, y_val)
print(f'Validation Accuracy: {accuracy}')

# Make predictions on the test data
x_test = test_space_d[features]
predictions = model.predict(x_test)

# Add predicted values as 'Transported' column to test_space_d
test_space_d['Transported'] = predictions

# Print or further analyze test_space_d with predictions
print(test_space_d)

```

```python
from sklearn.metrics import confusion_matrix

# Assuming model is trained and x_val, y_val are defined
predictions = model.predict(x_val)
cm = confusion_matrix(y_val, predictions)

print("Confusion Matrix:")
print(cm)
```

```python
from sklearn.metrics import classification_report

# Assuming model is trained and x_val, y_val are defined
predictions = model.predict(x_val)

# Print classification report
print(classification_report(y_val, predictions))
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Make predictions on the validation set
predictions = model.predict(x_val)

# Compute confusion matrix
cm = confusion_matrix(y_val, predictions)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
